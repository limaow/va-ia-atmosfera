```python
# Bibliotecas e manipulações

import pandas as pd
from pandas import read_csv
import plotly
import plotly.io as pio
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Colocando os dados, que estão no github, na variável atmosfera para manipular

atmosfera = pd.read_csv("https://raw.githubusercontent.com/limaow/datasets/main/atmosfera.csv")
```

```python
# Excluindo as colunas desnecessárias para formação da matriz de conhecimento

atmosfera = atmosfera.drop(columns=['DATA', 'temperatura'])
```

```python
# Aqui fica a "previsão" de dados pelo modelo com janela deslizante de razão 5

# Valores da coluna "umidade" em uma lista para manipular
valores = list(atmosfera.umidade)                                                         

tamanhoJanela = 5

# Executando uma lista de listas onde cada elemento será uma lista
# para armazenar os rótulos das colunas da janela deslizante                                                           
linhasJanela = []                                                                         
colunas = []                                                                              

for i in range(tamanhoJanela - 1):
    colunas.append(str(i + 1))

colunas.append("Valor esperado")

i = 0
# Estrutura While-not- para que a janela deslizante não consiga exibir 
# uma linha menor que seu tamanho

while not i + tamanhoJanela > len(valores) - 1:

# Lista que ira armazenar a linha da janela que será gerada                                          
    linhaI = []                                                                           

# Gerando linha da janela e salvando na lista linhaI criada acima
    for j in range(i, i + tamanhoJanela):                                                 
        linhaI.append(valores[j])

# Salvando a linha gerada em uma lista
    linhasJanela.append(linhaI)                                                           
    i = i + 1

# Transformando a lista de linhas em um array numpy np
linhasJanela = np.array(linhasJanela)

# Gerando e exibindo o DataFrame da janela deslizante
dataFrame = pd.DataFrame(data=linhasJanela, columns=colunas)                             
dataFrame
```
