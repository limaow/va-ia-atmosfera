```python
catatmosfera = '/content/atmosfera.csv'
atmosfera = read_csv(catatmosfera)
```

```python
atmosfera.head(10)
```

```python
atmosfera = atmosfera.drop(columns=['DATA', 'temperatura'])
```

```python
valores = list(atmosfera.umidade)                                                         

tamanhoJanela = 5                                                                         
linhasJanela = []                                                                         
colunas = []                                                                              

for i in range(tamanhoJanela - 1):
    colunas.append(str(i + 1))

colunas.append("Valor esperado")

i = 0
while not i + tamanhoJanela > len(valores) - 1:                                           
    linhaI = []                                                                           

    for j in range(i, i + tamanhoJanela):                                                 
        linhaI.append(valores[j])
    
    linhasJanela.append(linhaI)                                                           
    i = i + 1

linhasJanela = np.array(linhasJanela)                                                    
dataFrame = pd.DataFrame(data=linhasJanela, columns=colunas)                             

dataFrame
```
